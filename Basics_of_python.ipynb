# prime number code


```python
def is_prime(n):

 if n <= 1:
    return False


 for i in range(2, int(n**0.5) + 1):
  if n % i == 0:
    return False

 return True

num = int(input('Enter a number:'))

if is_prime(num):
  print(f'{num} number is prime')
else:
  print(f'{num} number is not prime')
```

Product of Random Numbers

```python
import random
def product():

  num1= random.randint(1,10)
  num2= random.randint(1,10)

  correct_product= num1*num2
  print(f"what is is the product of {num1} and {num2}")

  user_product=int(input('in the answer'))

  if correct_product == user_product:
    print('right answer')
  else:
    print('wrong answer')
    print(f'the correct answer is {correct_product}')

product()
```

Squares of Even/Odd Numbers



```python
def squares():

  choice=input('enter even or odd')
  if choice == 'even':
    for num in range(100, 201):
      if num % 2 == 0:

       print(f'number{num}, square{num**2}')

  elif choice == 'odd':
    for num in range(100, 201):
      if num % 2 != 0:

       print(f'number{num}, square{num**2}')
  else:
    print('invalid choice')

squares()
```

Hospital Billing System

```python
def bill():
  room=int(input('how many days '))
  doctor_fee=1500
  lab=int(input('how many lab test '))
  medicine=int(input('how many medicine charge '))

  room_charges=2000*room
  lab_charges=300*lab


  total=room_charges+doctor_fee+lab_charges+medicine

  if total > 10000:
    discount = total*0.10
    after_discount = total - discount
    print(f'your total bill is {after_discount}')
  else:
    print(f'your total bill is {total}')



bill()
```
