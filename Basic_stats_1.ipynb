Descriptive Analytics for Numerical Columns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
# Average of all data point in columns is know as mean
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].mean()
```

```python
# middle number of the dataset when is arranged in ascending order it know as median
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].median()
```

```python
# value come so many time in the dataset it know as mode
df[['Volume','Avg Price','Total Sales Value']].mode()
```

```python
# Standand deviation
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].std()
```

Data Visualization

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 3))
axes[0].hist(df['Volume'])
axes[0].set_title('Volume')
axes[1].hist( df['Avg Price'])
axes[1].set_title('Avg Price')
axes[2].hist( df['Total Sales Value'])
axes[2].set_title('Total Sales Value')


plt.show()


# all histogra are right skewness
```
